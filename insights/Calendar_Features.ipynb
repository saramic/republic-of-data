#### FE idea
 - Weekday - Done
 - Weekend - Done
 - Puplic holiday during weekday - Done
 - Puplic holiday during weekend - Done
 - School holiday or public holiday - Done
 - Day before/after any holiday - Done
 - Day before/after public holiday - Done
 - Day before/after scholl holiday
 - Day before/after weekend
 - Is Long Weekend (3 days continous holiday)
 - AL Long Weekend (public holiday on Tuesday or Thrusday)


```python
## FE Code

df_calendar['isWeekday'] = df_calendar.Date.dt.dayofweek >= 5
df_calendar['isWeekend'] = df_calendar.Date.dt.dayofweek < 5
df_calendar['iPH'] = df_calendar.DayType == 'Public Holiday'
df_calendar['iSchoolH'] = df_calendar.WeekdaySeq == 'School Holiday Weekday'
df_calendar['isWeekday_PH'] = df_calendar.isWeekday & (df_calendar.iPH)
df_calendar['isWeekend_PH'] = df_calendar.isWeekend & (df_calendar.iPH)
df_calendar['isAnyHoliday'] = df_calendar.isWeekend & ((df_calendar.iPH).astype(bool) | 
                                                       (df_calendar.iSchoolH).astype(bool))
df_calendar['isWeekdayBeforeHoliday'] = df_calendar.shift(1).isAnyHoliday & df_calendar.isWeekday
df_calendar['isWeekdayAfterHoliday'] = df_calendar.shift(-1).isAnyHoliday & df_calendar.isWeekday
df_calendar['isWeekdayBeforePH'] = df_calendar.shift(1).iPH & df_calendar.isWeekday
df_calendar['isWeekdayAfterPH'] = df_calendar.shift(-1).iPH & df_calendar.isWeekday

# Friday
df_calendar['isBeforeWeekend'] = df_calendar.Date.dt.dayofweek == 4 

# Monday
df_calendar['isBeforeWeekend'] = df_calendar.Date.dt.dayofweek == 0

# Is Long Weekday


df_calendar['isLongWeekend'] = (df_calendar.isBeforeWeekend & df_calendar.isWeekday.shift(-1)
                                (df_calendar.isWeekday.shift(-1) | ) | 
                                df_calendar.isWeekday.shift(-2))

```
